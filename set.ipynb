## ***Implementation of set***

- **Creating a Set**

```python
my_set = {10, 20, 30, 40, 50}
print("Original Set:", my_set)
```

- **Adding an Element**

```python
my_set.add(60)
print("After Adding 60:", my_set)
```

- **Adding Multiple Elements**

```python
my_set.update([70, 80])
print("After Updating:", my_set)
```

- **Removing an Element**

```python
my_set.remove(30)
print("After Removing 30:", my_set)
```

- **Discard an Element**

```python
my_set.discard(20)
print("After Discarding 20:", my_set)
```

- **Traversing the Set**

```python
print("\nTraversing Set:")
for i in my_set:
    print(i)
```

- **Length of Set**

```python
print("Length of Set:", len(my_set))
```

- **Checking Membership**

```python
if 40 in my_set:
    print("40 is present in the set")
```

- **Pop an Element**

```python
my_set.pop()
print("After Pop:", my_set)
```

- **Clearing the Set**

```python
temp_set = {1, 2, 3}
temp_set.clear()
print("Before Clearing:", temp_set)
print("Cleared Set:", temp_set)
```

- **Iterate Through Set**

- **Union of Sets**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

- **Intersection of Sets**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

```python
a = {1, 2, 3}
b = {2, 3}

print(a.difference(b))
```

```python
a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b))
```
