### ***Implementation of List in Python***

- **Creating a List**

```python
my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)
```

- **Accessing Elements**

```python
print("First Element:", my_list[0])
print("Last Element:", my_list[-1])
```

- **Traversing the List**

```python
print("\nTraversing List:")
for item in my_list:
    print(item)
```

-  **Adding Elements**

```python
my_list.append(60)     # Add at end
print("\nAfter Append:", my_list)
```

```python
my_list.insert(2, 25)       # Insert at specific index
print("After Insert at index 2:", my_list)
```

- **Removing Elements**

```python
my_list.remove(30)          # Remove specific value
print("\nAfter Remove 30:", my_list)
```

```python
popped = my_list.pop()      # Remove last element
print("After Pop:", my_list)
print("Popped Element:", popped)
```

- **Updating Elements**

```python
my_list[1] = 200
print("\nAfter Updating index 1:", my_list)
```

- **Length of List**

```python
print("\nLength of List:", len(my_list))
```

- **Sorting**

```python
my_list.sort()
print("Sorted List:", my_list)
```

- **Reversing**

```python
my_list.reverse()
print("Reversed List:", my_list)
```

- **Slicing**

```python
if 40 in my_list:
    print("40 is present in the list")
else:
    print("40 is not present in the list")
```

- **List Comprehension**

```python
# Square of numbers from 1 to 100
squares = [x**2 for x in range(1, 101)]
print("Squares:", squares)
```

- **Even Numbers from List**

```python
numbers = [10, 15, 20, 25, 30, 35]
even = [num for num in numbers if num % 2 == 0]
print("Even Numbers:", even)
```

- **Find Maximum and Minimum**

```python
numbers = [45, 12, 89, 33, 67]
print("Maximum number is :", max(numbers))
print("Minimum number is :", min(numbers))
```

- **Sum of List Elements**

```python
numbers = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 121, 144, 169]
total = sum(numbers)
print("Sum:", total)
```

- **Count Occurrence of Element**

```python
numbers = [1, 2, 2, 3, 4, 2, 5]

print("Count of 2:", numbers.count(2))
```

- **Copy a List**

```python
list1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 121, 144, 169, 196, 225, 256, 289, 324, 361, 400, 441, 484, 529, 576, 625, 676, 729, 784, 841, 900, 961]
list2 = list1.copy()
print("Original List:", list1)
print("Copied List:", list2)
```

- **Merge Two Lists**

```python
list1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 121, 144, 169, 196, 225, 256, 289, 324, 361, 400, 441, 484, 529, 576, 625, 676, 729, 784, 841, 900, 961, 1024]
list2 = [1089, 1156, 1225, 1296, 1369, 1444, 1521, 1600, 1681, 1764, 1849, 1936, 2025, 2116, 2209, 2304, 2401, 2500, 2601, 2704, 2809, 2916, 3025, 3136, 3249, 3364, 3481, 3600, 3721, 3844, 3969, 4096, 4225, 4356, 4489, 4624, 4761, 4900, 5041, 5184, 5329, 5476, 5625, 5776, 5929, 6084, 6241]

merged = list1 + list2
print("Oribginal List 1:", list1, "\nOriginal List 2:", list2)
print("Merged List:", merged)
```

- **Remove Duplicates from List**

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique = list(set(numbers))
print("Original List with Duplicates:", numbers)
print("Without Duplicates:", unique)
```

- **Reverse List using Slicing**

```python
numbers = [1, 2, 3, 4, 5]

reversed_list = numbers[::-1]
print("Reversed List:", reversed_list)
```

- **Nested List (2D List)**

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrix:")
for row in matrix:
    print(row)
```

- **Flatten a Nested List**

```python
matrix = [[1, 2], [3, 4], [5, 6]]

flat = [item for sublist in matrix for item in sublist]
print("Flattened List:", flat)
```

- **Sort List in Descending Order**

```python
numbers = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 121, 144, 169, 196, 225, 625, 676, 729, 784, 841, 900, 256, 289, 324, 361, 400, 441, 484, 529, 576, 961, 1024]

numbers.sort(reverse=True)
print("Descending Order:", numbers)
```

- Find Second Largest Number

```python
numbers = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 121, 144, 169, 196, 225, 625, 676, 729, 784, 841, 900, 256, 289, 324, 361, 400, 441, 484, 529, 576, 961, 1024]

numbers.sort()
print("Second Largest:", numbers[-2])
```

- **Check if List is Empty**

```python
my_list = []

if not my_list:
    print("List is Empty")
else:
    print("List is Not Empty")
```

```python
text = "CHINNASWAMIMUTTHUSWAMIVENUGOPALIYER"

char_list = list(text)
print("String to List:", char_list)
```
