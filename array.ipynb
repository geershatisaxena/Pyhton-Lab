### ***Implementation of Array using List***

-  **Creating an array (using list)**

```python
arr = [10, 20, 30, 40, 50]
```

 - **Accessing elements**

```python
print("First element:", arr[0])
print("Last element:", arr[-1])
```

- **Traversing array**

```python
print("Array elements:")
for i in arr:
    print(i)
```

- **Inserting element**

```python
arr.append(60)
print("After insertion:", arr)
```

- **Inserting at specific position**

```python
arr.insert(2, 25)
print("After inserting 25 at index 2:", arr)
```

- **Deleting element**

```python
arr.remove(30)
print("After removing 30:", arr)
```

- **Updating element**

```python
arr[1] = 200
print("After updating index 1:", arr)
```

- **Length of array**

```python
print("Length of array:", len(arr))
```

### ***Implementation using array Module***

- **Create array of integers**

```python
import array
arr = array.array('i', [1, 2, 3, 4, 5])

print("Original Array:", arr)
```

- **Access element**

```python
print("Element at index 2:", arr[2])
```

- **Append element**

```python
arr.append(6)
print("After append:", arr)
```

- **Insert element**

```python
arr.insert(1, 10)
print("After insert:", arr)
```

- **Remove element**

```python
arr.remove(3)
print("After remove:", arr)
```

- **Traverse array**

```python
print("Array elements:")
for i in arr:
    print(i)
```

### ***Implementation using NumPy Array (Advanced)***

- **Creating NumPy array**

```python
import numpy as np
arr = np.array([5, 10, 15, 20])
print("Numpy array:", arr)
```

- **Accessing element**

```python
print("First element:", arr[0])
```

- **Array operations**

```python
print("Array + 5:", arr + 5)
```

- **2D Array**

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)
```
