### ***Implementation of Tuple in Python***

- **Creating a Tuple**

```python
my_tuple = (10, 20, 30, 40, 50)
print("Original Tuple:", my_tuple)
```

- **Accessing Elements**

```python
print("First Element:", my_tuple[0])
print("Last Element:", my_tuple[-1])
```

- **Traversing the Tuple**

```python
print("\nTraversing Tuple:")
for item in my_tuple:
    print(item)
```

- **Length of Tuple**

```python
print("\nLength of Tuple:", len(my_tuple))
```

- **Counting an Element**

```python
print("Count of 20:", my_tuple.count(20))
```

- **Finding Index**


```python
print("Index of 30:", my_tuple.index(30))
```

- **Tuple Slicing**

```python
print("Tuple Slice (1 to 3):", my_tuple[1:4])
```

- **Concatenation of Tuples**

```python
tuple2 = (60, 70)
new_tuple = my_tuple + tuple2
print("Before Concatenation:", my_tuple)
print("After Concatenation:", new_tuple)
```

- **Repetition**

```python
repeat_tuple = my_tuple * 2
print("Repeated Tuple:", repeat_tuple)
```

- **Membership Test**

```python
if 40 in my_tuple:
    print("40 exists in tuple")
else:
    print("40 not found")
```

- **Nested Tuple**

```python
nested_tuple = ((1, 2), (3, 4), (5, 6))
print("\nNested Tuple:", nested_tuple)
```

- **Create Tuple with Different Data Types**

```python
t = (10, "Python", 3.14, True)
print(t)
```

- **Single Element Tuple**

```python
t = (5,)
print(t)
print(type(t))
```

- **Convert List to Tuple**

```python
lst = [1, 2, 3, 4]
t = tuple(lst)
print(t)
```

- **Convert Tuple to List**

```python
t = (10, 20, 30)
lst = list(t)
print(lst)
```

- **Access Tuple Elements**

```python
t = (100, 200, 300)
print(t[1])
```

- **Negative Indexing**

```python
t = (10, 20, 30, 40)
print(t[-1])
```

- **Tuple Slicing**

```python
t = (1, 2, 3, 4, 5)
print(t[1:4])
```

- **Iterate Through Tuple**

```python
t = (5, 10, 15, 20)

for i in t:
    print(i)
    
```

- **Check Element in Tuple**

```python
t = (10, 20, 30)

if 20 in t:
    print("Element Found")
```

- **Length of Tuple**

```python
t = (1, 2, 3, 4, 5)
print(len(t))
```

- **Tuple Concatenation**

```python
t1 = (1, 2)
t2 = (3, 4)

t3 = t1 + t2
print(t3)
```

- **Tuple Repetition**

```python
t = (1, 2, 3)
print(t * 3)
```

- **Count Elements**

```python
t = (1, 2, 2, 3, 2)
print(t.count(2))
```

- **Find Index**

```python
t = (10, 20, 30)
print(t.index(20))
```

- **Maximum and Minimum**

```python
t = (5, 8, 2, 10)
print("Max:", max(t))
print("Min:", min(t))
```

- **Sum of Tuple Elements**

```python
t = (1, 2, 3, 4)
print("Sum:", sum(t))
```

- **Nested Tuple**

```python
t = ((1, 2), (3, 4), (5, 6))
print(t[1])
```

- **Unpacking Tuple**

```python
t = (10, 20, 30)
a, b, c = t
print(a)
print(b)
print(c)
```

- **Delete Entire Tuple**

```python
t = (1, 2, 3)

del t
print("Tuple Deleted")
```
