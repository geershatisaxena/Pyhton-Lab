### ***Implementation of Tuple in Python***

- **Creating a Tuple**

```python
my_tuple = (10, 20, 30, 40, 50)
print("Original Tuple:", my_tuple)
```

- **Accessing Elements**

```python
print("First Element:", my_tuple[0])
print("Last Element:", my_tuple[-1])
```

- **Traversing the Tuple**

```python
print("\nTraversing Tuple:")
for item in my_tuple:
    print(item)
```

- **Length of Tuple**

```python
print("\nLength of Tuple:", len(my_tuple))
```

- **Counting an Element**

```python
print("Count of 20:", my_tuple.count(20))
```

- **Finding Index**


```python
print("Index of 30:", my_tuple.index(30))
```

- **Tuple Slicing**

```python
print("Tuple Slice (1 to 3):", my_tuple[1:4])
```

- **Concatenation of Tuples**

```python
tuple2 = (60, 70)
new_tuple = my_tuple + tuple2
print("Before Concatenation:", my_tuple)
print("After Concatenation:", new_tuple)
```

- **Repetition**

```python
repeat_tuple = my_tuple * 2
print("Repeated Tuple:", repeat_tuple)
```

- **Membership Test**

```python
if 40 in my_tuple:
    print("40 exists in tuple")
else:
    print("40 not found")
```

- **Nested Tuple**

```python
nested_tuple = ((1, 2), (3, 4), (5, 6))
print("\nNested Tuple:", nested_tuple)
```
